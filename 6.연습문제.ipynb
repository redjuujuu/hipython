```python
#ì§€í”¼í‹°

def check_all_digits_once(s):
    # ê¸¸ì´ê°€ 10ì´ ì•„ë‹ˆë©´ ë°”ë¡œ False
    if len(s) != 10:
        return False
    
    # 0~9ê°€ ê°ê° í•œ ë²ˆì”© ìˆëŠ”ì§€ í™•ì¸
    return set(s) == set('0123456789')


# í…ŒìŠ¤íŠ¸
inputs = [
    "0123456789",
    "01234",
    "01234567890",
    "6789012345",
    "012322456789"
]

for i in inputs:
    print(check_all_digits_once(i))
```
