```python
# 문자열 순회
password = 'abc21!'
isUpper = False
isDigit = False
special_char = '!@#%?$'
isSpecial = False
isLower = False
#비밀번호 조건 만족여부 확인
for char in password :
    if char.isupper() : #대문자인지 확인
        isUpper = True
    elif char.isdigit() : #숫자인지 확인
        isDigit = True

    elif char.islower() : #소문자인지 확인
        isLower = True
    elif char in special_char :
        isSpecial = True
    check_result = isUpper + isDigit + isLower + isSpecial   #and는 이렇게 숫자를 기준으로 해서 하는게 나음
print(f"'비밀번호 강도 조건 만족여부 : {'만족' if check_result == 4 and isDigit else '불만족'}")
```

```python
#연도 계산
total_days = 0

for i in range(1, year):      #여기서 (1,year)은 뭐야  #range가 빠져있었음   #2024년까지
    if i % 400 == 0 : #윤년
        total_days += 366
    elif i % 100 == 0 : #평년     #예외규칙이니까 먼저
        total_days += 365
    elif i % 4 == 0 : #윤년
        total_days += 366
    else : #평년
        total_days += 365  

print(f'1st : {total_days}, {year}')


# 2024년의 월기준 일수 계산
    #처리해주기 위해서 리스트 만들것임

month_days = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

for m in range(1,month):  #1안쓰면 0부터 돌림
    total_days += month_days[m]    #이게 뭔말ㄹ이고   #total_days += month_days[month]  이랬어서 잘못나옴

print(f'2nd : {total_days}, {year}')   #(앞에거에 비해 59일 더하는...)
#윤년의 규칙도 추가해줘야함 - 4로 나누어떨어지면 윤년 / 100으로 나누어떨어지면 윤년아님 / 400으로 떨어지면 윤년


#올해가 윤년인지 체크해보고 맞으면 +1해주면 됨
if month >= 3 :
    if year % 400 == 0 : #윤년
        total_days += 1
    elif year % 100 == 0 : #평년     #예외규칙이니까 먼저
        total_days += 0
    elif year % 4 == 0 : #윤년
        total_days += 1
    else : #평년
        total_days += 0   #위에있는거 복사해서 숫자만 다르게함

print(f'3nd : {total_days}, {year}, {month}')

total_days += day

print(f'final : {total_days}, {year}, {month}, {day}')

#연도 + 월 + 일 %7 > 요일에 해당하는 수

remain = total_days % 7
result = None

if remain == 0:
    result = '일요일'
elif remain == 1:
    result = '월요일'
elif remain == 2:
    result = '화요일'
elif remain == 3:
    result = '수요일'
elif remain == 4:
    result = '목요일'
elif remain == 5:
    result = '금요일'
elif remain == 6:
    result = '토요일'




# 위에 month 쓴 방식대로 할 수도 있음
month_days = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

for m in range(1,month):  #1안쓰면 0부터 돌림
    total_days += month_days[m] 

print(f'2nd : {total_days}, {year}')  


#내 방식.... 딕셔너리로 시작함

week_day = { '0': '일', '1': '월', '2' :'화', '3' : '수', '4': '목', '5' :'금', '6': '토'}

print(f' {year}년 {month}월 {day}일은 {result}입니다.')
```

```python
#정답 맞추기
cnt = 0             #이거 왜 하는거야?
while True:
    #입력받기
    answer = int(input('1~10 사이의 숫자를 맞추세요'))
    cnt += 1                    #입력하는 순간 숫자가 늘어나야 하니까
    #조건
    if answer == number : #맞춘 경우
        print(f'축하합니다. 정답입니다. 도전횟수 : {cnt}')
        break
    elif answer > number :
        print('down')
    else :
        print('up')
        
```
