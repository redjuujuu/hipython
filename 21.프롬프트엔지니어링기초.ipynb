# 헬퍼 - input_text 유형의 메시지 처리

````python
def msg(role, text) :
    msg_src = {"role": role, "content": [{"type": "input_text", "text":text}] }   #openai가 이름도 자꾸 바꾸고 response 구조도 바꾸니까
    return msg_src
````

- system

- assistant : 그 전에 기억하고 싶은 대화를 저장 -> 래그로 간다


````python
res = client.responses.create(
    model = 'gpt-4o-mini',        #파라미터와 파라미터는 콤마로 연결!!
    input = [                       #리스트 구조로 연결
        msg("user", '2002년 월드컵에서 가장 화제가 되었던 나라는 어디야?'),
        {
            "role":"assistant", #헬퍼는 그대로 두고, assistant만 명시적으로 처리
         "content":[     
             {
                 "type":"output_text", 
                 "text":"바로 예상을 뚫고 우승한 브라질입니다."
             }
        ]  
    },
    msg("user", ''' 그 나라가 화제가 되었던 이유를 자세하게 설명해줘''')  #이게 왜 계속 기억되는 원리인거야?
    ]
)
````

# 파라미터의 제어
- temperature : 창의성 0~1 (최대)
- max_output_tokens : 최대 응답 길이
- top_p : 후보선정 기준 설정
- frequency_penalty : 반복 억제

````python
top_p = [0.6, 0.8, 1.0]
for p in top_p:
    response = client.responses.create(
        model='gpt-4o-mini',
        input='마케팅 문구 5개 만들어줘',
        #temperature=0.5,
        top_p=p
)
print(response.output_text)
````
