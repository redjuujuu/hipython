```python
import folium
from math import radians, sin, cos, sqrt, atan2

# ✅ 1) 좌표 입력 (위도, 경도)
home = (37.5665, 126.9780)     # 예시: 서울시청 근처
academy = (37.5700, 126.9920)  # 예시: 종로 쪽


# ✅ 3) 지도 생성 (두 점의 중간으로 중심 맞추기)
center = ((home[0] + academy[0]) / 2, (home[1] + academy[1]) / 2)
m = folium.Map(location=center, zoom_start=15)

# ✅ 4) 마커
folium.Marker(
    location=home,
    popup="집",
    icon=folium.Icon(color="blue", icon="home")
).add_to(m)

folium.Marker(
    location=academy,
    popup="학원",
    icon=folium.Icon(color="red", icon="flag")
).add_to(m)

# ✅ 5) 직선(폴리라인) + 거리 표시(툴팁)
folium.PolyLine(
    locations=[home, academy],
    weight=5,
    opacity=0.8,

).add_to(m)


# ✅ 7) 저장
m.save("home_to_academy_straight_line.html")


m3 = folium.Map(location=((home[0]+academy[0])/2, (home[1]+academy[1])/2), zoom_start=15)
```
