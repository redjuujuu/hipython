# 상대도수

```python
import random
```

```python
exp_p = 0.5 #확률
trial_count = 10 #시행
head_count = 0 #앞면이 나온 건수   #얘는 왜 0으로 돼있지?
#동전의 앞면이 나올 확률 50%
#10번 시행했을 때 상대도수와 확률 비교

for cnt in range(trial_count):
    if random.random() < exp_p :
        head_count +=1

        
print(f'앞면이 나옷 횟수 : {head_count} ')
actual_freq = head_count/trial_count

print(f'예상확률 : {exp_p*100}%')
print(f'실제결과(상대도수) : {actual_freq*100}%')
```

# 확률

```python
import math
```

```python
total_balls =7
white_balls=4
black_balls=3
pick_count=3

#전체 경우의 수
total_case = math.comb(total_balls, pick_count)

#흰 공 2개 나올 경우의수 (4,2)
white_case = math.comb(white_balls, 2)

#검은 공 1개 나올 경우의수 (3,1)
black_case = math.comb(black_balls, 1)

#흰공2, 검은공1 경우의 수
event_case = white_case * black_case

prob = event_case / total_case

print(f' 계산한 확률 : {prob*100:.2f}')
```

```python
import random
```

```python
#시뮬레이션

def run_simulation(trials):
    act_freq = 0
    success_cnt = 0
    sample_space =['W','W','W','W','B','B','B']   #표본공간
    for i in range(trials):                 #여기서 왜 range야?
        drawn_balls = random.sample(sample_space, 3)
        if drawn_balls.count('W') == 2 and drawn_balls.count('B') == 1:   #elif는 안되나?
            success_cnt +=1 #흰공2, 검공1개나온 횟수 
    act_freq = success_cnt / trials #상대도수
    return act_freq
```

```python
trials = 10
simulation_prob = run_simulation(trials)
print(f'시뮬레이션횟수 {trials}, 결과확률 {simulation_prob*100:.2f}')
```

# 확률의 법칙

```python
# 독립사건 확률의 곱셈법칙 : 주사위와 동전 - 확률가지고 함

import random

a_prob = 0.5 #주사위 짝수
b_prob = 0.5 #동전 앞면
prob_ind = a_prob * b_prob
print(f'독립사건 확률이 곱 : {prob_ind:.2f}')
```

```python
#종속사건
total = 10
win = 2
prob_A = win / total #A가 당첨될 확률 P(A)
prob_B = (win-1) / (total-1) #A 다음에 B가 당첨될 확률 P(B|A)
prob_AB = prob_A * prob_B #A확률 B확률의 곱 = P(A) * P(B|A)
print(f'종속사건 AB의 확률 : {prob_AB:.2f}')
```
