```python
#노션 조건문 예제 - 특정단어가 문장에 포함되어 있는지 확인하는 프로그램 
#입력된 문장은 sentence 변수에 저장하며, 검사할 단어는 리스트 words에 저장한다

sentence = input('문장을 입력하세요:')
words = ['파이썬', '코딩', '데이터', 'AI']

found_word = None   #얘는 값이 없다는 예약어

for word in words:
    if word in sentence:
        found_word = word
        break

if found_word:
    print(f'입력한 문장에서 "{found_word}" 단어를 찾았습니다!')
else:
    print('입력한 문장에 해당 단어가 없습니다.')
```

# break/continue/pass 

```python
numbers = [1, 3, 5, 7, 9, 11, 13, 15]
s_value = 9

for n in numbers : 
    if (n == s_value) :
        print(f'찾았습니다. {n}')
        break    #가장 가까운 반복문만 끝냄
    else : 
        print( f'현재 검사한 숫자는 : {n}')
```
