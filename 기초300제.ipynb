;세미콜론은 명령을 어떻게 써놓느냐의 문제이지,
무엇이 출력되는지(결과)를 바꾸는 문법이 아님.

세미콜론의 역할 -> 한 문장, 한 명령이 끝났다고 알려주는 것 - 작성과 해석의 기준임
세미콜론은 줄바꿈을 대체할 뿐
