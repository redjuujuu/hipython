# 인덱싱
[] 인덱싱

- iloc[]행기반 인덱싱 >>> [행, 열]
- loc[] 이름기반(위치) 인덱싱 >> [행이름, 열이름]

- :의 의미 >> start:end 까지 슬라이싱
- loc[행이름:행이름]
- iloc[행번호:행번호] >> end 행번호는  -1까지


```python
df1
```

```python
#df1[['이름', 'age']]  #여러개 할 때는 리스트로 만들어서!!!

df1.loc[  : , ['이름', 'age']] #행과 열을 지정해줘야함
```

```python
df1.loc['A'] #행, 열
```

```python
df1.iloc[0:3, 0:2] #행번호, 열, :은 start:end(-1)
```

```python
# # 연습문제
# [기본]

# 모든 행에서 이름 열만 선택하세요.
df1
df1.loc[ : ,'이름']

# 모든 행에서 이름과 score 열을 선택하세요.
df1.loc[ : , ['이름', 'score']]

# 인덱스 C 행의 모든 데이터를 선택하세요.
df1.loc['C']


# 인덱스 D 행의 age 값만 선택하세요.
df1.iloc[3:4, 1:2]
```

```python
# [슬라이싱]


# 인덱스 B부터 E까지의 모든 행을 선택하세요.
df1.loc['B': 'E']

# 인덱스 C부터 F까지의 이름과 age 열만 선택하세요.
df1.loc['C':'F', '이름':'age']

# 인덱스 A부터 G까지에서 score 열만 선택하세요.
df1.loc['A':'G', 'score']
```

# boolean 인덱싱

```python
df[df['score'] > 90]  #시리즈의 모든 요소에 적용 - #vectorization - boolean list의 요소값 True, False로 해놓고 그 결과를 나타낸 것
```

```python
#[조건 + loc]
#age가 27 이상인 행만 선택하세요.
df1.loc[df1['age']>=27]
cond = df1['age']>= 27
df1.loc[cond]

#score가 90 이상인 사람의 이름과 score만 출력하세요.
#df1.loc[df1['score']>=90] - 아 age가 나오는구나
#df1.loc[행, 열]

df1.loc[df1['score']>=90 , ['이름', 'score']]

#age가 25 이상이면서 score가 90 이상인 행을 모두 선택하세요.
#df1.loc[df1.['age']>=25 and df1.['score']>=90]

cond1 = df1['age'] >=25
cond2 = df1['score']>=90
df1.loc[cond1 & cond2]
```

```python
# [기본]

# 첫 번째 행의 모든 데이터를 선택하세요.

# 세 번째 행의 이름 값만 선택하세요.

# 처음 3개의 행을 선택하세요.

# 마지막 행의 score 값만 선택하세요.
df1.iloc[-1,2]
```

```python
# [슬라이싱]

# 1번 행부터 4번 행까지 선택하세요.
df1.iloc[1:5]

# 2번 행부터 5번 행까지에서 이름과 age 열만 선택하세요.
df1.iloc[2:6, 0:2]

# 모든 행에서 age 열만 선택하세요.
df1.iloc[ : , 1] # 콤마 앞은 생략하면 안됨 - 처음부터 끝까지는 콜론으로 표기
```

```python
# [조합 문제]

# score 열만 선택하되, 앞에서 4개의 행만 출력하세요.
    # df1.iloc[:4, 'score'] -> 이러면 에러남 - score라는 이름 모른다면서..
df1.iloc[:4, 2]

# 이름과 score 열을 선택하되, 마지막 3개의 행만 출력하세요.
df1.iloc[-3:, ::2 ]     #-3: 인 이유? :-3이아니라? ::2는 뭐였지?

    #이렇게도 가능
df1.tail(3).loc[:,['이름', 'score']] 
df1.iloc[-3: , [0,2]]

# 짝수 번째 행(0, 2, 4, 6)의 모든 데이터를 선택하세요.
df1.iloc[[0,2,4,6]] #아 그냥 선택하면 되는구나
```
