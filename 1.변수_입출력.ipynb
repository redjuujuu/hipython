#변수의 영역 namespace

지역변수, 전역변수

```python
def local_exam(): #사용자정의함수
    local_var = 100   #지역변수 = local variable
    print(local_var)

local_exam()
local_var = 200   #전역변수
local_exam()
print(local_var)
```

```python
def local_exam(): 
    local_var = 100   
    print(f"함수내부: {local_var}")   #얘는 정의만 내린것임

local_exam()    # 얘가 호출한것 - 호출하는 순간 진행이됨 1,2,3은 기억만함
local_var = 200   #local_var을 다르게 해도 상관이 없음
local_exam()
print(f"함수외부: {local_var}") #5,7.8이 결과로 나온것
```

```python
counter = 0 #전역변수
def inc_counter():
    global counter
    counter = counter + 1  #앞의 counter는 지역변수  
    print(counter)

#temp = a + 1   프로그래밍에서는 한줄로 표현 가능함
#a = temp 이렇게 하지 않아도
#a = a + 1 이렇게 프로그래밍에서 할 수 있음

a += 1 #9번이랑 11번이랑 같은 말임 / 코딩에서는 이렇게 하는거야~ 하다가
#사고가 다르니까 자체적으로 줄이는 구나 생각하면 될 듯

inc_counter()
inc_counter()
inc_counter()
```
