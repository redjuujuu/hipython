변수 price와 is_discount가 주어졌을 때, is_discount가 True이면 가격에 10% 할인을 적용하고, 그렇지 않으면 원래 가격을 출력하는 코드를 작성하세요.

```python
#내버젼
price = int(input('가격을 입력하세요'))
is_discount = True
new_price = 0.9 * price

if is_discount == True:
    print (f' 가격은 {new_price}원 입니다.')
    
else:
    print (f' 가격은 {price}원 입니다.')
```

```python
vip_list = ['조인성', '김우빈', '정우성', '유해진']
num=1
for vip in vip_list:
    #print(vip)
    print(f'{vip}님 안녕하세요?')
    print(f'화창한 2023년 가을, 저희 컴퓨터 동아리 축제에 {vip}님을 초대합니다.')
    print(f'쿠폰 No.{num}')
    num +=1
    print(f'{vip}님 축제에서만나길 고대하겠습니다.')



#vip 목록을 입력받는걸로 바꿔보기
#enumerate로도 해보기
#####들여쓰기 까먹지말기!!!
```

```python
#vip 목록을 입력받는걸로 바꿔보기
vip_list = list(input( '성함을 입력해주세요').split[' '])
num=1
for vip in vip_list:
    #print(vip)
    print(f'{vip}님 안녕하세요?')
    print(f'화창한 2023년 가을, 저희 컴퓨터 동아리 축제에 {vip}님을 초대합니다.')
    print(f'쿠폰 No.{num}')
    num +=1
    print(f'{vip}님 축제에서만나길 고대하겠습니다.')




#enumerate로도 해보기
#####들여쓰기 까먹지말기!!!
```
