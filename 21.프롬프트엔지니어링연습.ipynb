````python
#메시지
sys_prompt = '''
너는 프로야구팀에서 근무하는 밈을 잘 아는 마케팅 전문가야
'''
user_prompt = '''
인스타그램 홍보 문구 3개 만들어줘'''
````

````python
def msg(role, text):
    
    return {"role":role, "content": text}  #return할 때 왜 딕셔너리로 해주면 돼?
````

````python
#프롬프트 생성
res = client.responses.create(   #()로해서 안됐음
    model = 'gpt-4o-mini'
    #input=[{"system":sys_prompt}, {"user":user_prompt}], #프롬프트가 두개니까 딕셔너리로
    , input=[{"role": "system", "content":sys_prompt}
    ,{"role": "user", "content":user_prompt}]
    , temperature=0.5

)
res.output_text

#매번 role, content하기 귀찮으니까 msg 헬퍼 함수 쓰기로함
````

````python
def format_gpt_output(text):
    """
    GPT 출력에서 \n을 실제 줄바꿈으로 변환하여 가독성 있게 출력
    """
    # \n을 실제 줄바꿈으로 변환
    formatted_text = text.replace('\\n', '\n')
    return formatted_text
````

````python
sys_prompt=f"""너는 {role}한 상사이다.
공격적이지 않으면서 설득력있게 전달해야 한다."""

user_prompt ="""직원에게 보내는 메일 초안을 작성하라
상황 : 항상 마감일 직전에 보고서를 제출하는 직원
"""

res = client.responses.create(
    model = 'gpt-4o-mini'
    , input = [msg("system", sys_prompt), msg("user", user_prompt)]
    , temperature = 0.4

)

print(format_gpt_output(res.output_text))
````

````python
roles =[ "잔소리꾼", 
    "유머러스", 
    "직원들의 태도에 화가난 ", 
    "냉정하고 규칙을 칼같이 지키는",
    "반말을 쓰는 무례한 상사"
]

for role in roles:
    sys_prompt = f""" 너는 {role}한 상사이다 공격적이지 않으면서 설득력있게 전달해야 한다.
    다만, 단순하고 직절석인 어투로 작성합니다."""
    res = client.responses.create(
        model='gpt-4o-mini',
        input=[msg("system", sys_prompt), msg("user", user_prompt)],
        temperature=0.9
    )
    print(f'{role}한 상사의 메일')
    print(format_gpt_output(res.output_text))
    print('*'*20)
````

````python
import os
import re
import requests
from bs4 import BeautifulSoup
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def msg(role, text):
    return {"role": role, "content": [{"type": "input_text", "text": text}]}

def fetch_url_text(url: str, timeout: int = 15) -> dict:
    """
    URL에서 텍스트를 추출해 반환.
    - title: 페이지 타이틀
    - text: 본문 텍스트(가능한 범위)
    - status: HTTP 상태코드
    - warnings: 추출 품질 경고
    """
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                      "(KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    }
    resp = requests.get(url, headers=headers, timeout=timeout)
    status = resp.status_code
    warnings = []

    if status != 200:
        return {"title": "", "text": "", "status": status, "warnings": [f"HTTP {status}"]}

    soup = BeautifulSoup(resp.text, "html.parser")

    # 불필요한 태그 제거
    for tag in soup(["script", "style", "noscript"]):
        tag.decompose()

    title = soup.title.get_text(" ", strip=True) if soup.title else ""

    # 메인 텍스트 후보: body 전체 텍스트
    raw_text = soup.get_text("\n", strip=True)

    # 너무 짧으면(동적 페이지/차단 가능성)
    if len(raw_text) < 500:
        warnings.append("추출된 텍스트가 매우 짧습니다(동적 페이지/차단/로그인 필요 가능).")

    # 공백 정리
    cleaned = re.sub(r"\n{3,}", "\n\n", raw_text).strip()

    return {"title": title, "text": cleaned, "status": status, "warnings": warnings}

def build_user_prompt_from_url(url: str, user_context: str = "") -> str:
    data = fetch_url_text(url)
    if data["warnings"]:
        warn_block = "\n".join([f"- {w}" for w in data["warnings"]])
    else:
        warn_block = "- 없음"

    # 너무 긴 텍스트는 토큰을 많이 먹으니 상한을 두는 게 안전(예: 12,000자)
    text_snippet = data["text"][:12000]

    prompt = f"""[상품 URL]
{url}

[페이지 제목]
{data["title"]}

[추출 텍스트(일부)]
{text_snippet}

[추출 경고]
{warn_block}

[사용자 추가 정보]
{user_context if user_context else "(미제공)"}
"""
    return prompt

# --- 사용 예시 ---
system_prompt = ''' 당신은 “충동구매 억제 코치”입니다. 목표는 사용자가 후회할 확률을 낮추고, 합리적인 소비 결정을 내리도록 돕는 것입니다.



## 핵심 규칙
- 5단계를 반드시 순서대로 평가합니다.
- 각 단계는 1~5점으로 채점합니다.
  - 1점에 가까울수록 “사면 안 되는 이유(충동/리스크↑)”
  - 5점에 가까울수록 “사도 되는 이유(필요/가치↑)”
- 총점 = (1~5단계 점수 합, 최대 25점)
- 결론 기준:
  - 총점 >= 20: 구매 권장
  - 총점 16~19: 보류(조건부 구매/대안 검토)
  - 총점 <= 15: 비구매 권장(대체/지연)

## 입력
사용자가 제공한 상품 정보(또는 URL에서 추출된 텍스트)와 사용자 상황(예산, 사용 목적, 보유 유사 제품, 구매 동기)을 사용합니다.
정보가 부족하면 “추가 질문”에 포함하여 요구합니다.

## 5단계 평가 및 채점 기준 (각 단계는 ‘판정 문장’과 ‘점수’를 반드시 포함)


### 1단계: 구매 이유의 명확성 및 비교 타당성 (충동구매 판별 핵심 단계)

이 단계에서는 “왜 이 상품을 사야 하는지”가 명확한지,
그리고 이미 보유한 제품과 비교했을 때 ‘실질적인 차이’가 있는지를 평가한다.
단순한 색상·디자인 변화나 기분 전환 목적은 충동구매 신호로 간주한다.

또한 리뷰 수와 리뷰 내용이 해당 구매 이유를 충분히 뒷받침하는지도 함께 고려한다.

- 충동 구매 신호에 해당하는 판정 문장 예시:
  - “이미 유사한 제품을 보유하고 있으며, 차이가 색상·디자인 등 외형적 요소에 그친다.”
  - “현재 사용 중인 제품에 큰 불편이나 기능적 한계가 명확히 드러나지 않는다.”
  - “구매 이유가 ‘갖고 싶다’, ‘예뻐서’, ‘기분 전환’ 중심이다.”
  - “리뷰 수가 매우 적거나, 구매 이유와 직접 관련된 리뷰 근거가 부족하다.”

- 합리적 구매 신호에 해당하는 판정 문장 예시:
  - “기존 보유 제품이 기능적·상태적 한계에 도달했다(노후, 고장, 체형 변화 등).”
  - “체형 변화(예: 살이 빠짐/늘어남), 사용 환경 변화 등으로 기존 제품이 더 이상 적합하지 않다.”
  - “새 제품이 기존 제품 대비 명확한 기능적 개선점(성능, 편의, 안전)을 제공한다.”
  - “리뷰 수가 충분히 많고, 구매 이유와 직결되는 긍정적 경험이 반복적으로 확인된다.” 

---

### 2단계: 필요성(Need vs Want) 판정
- 충동 신호 판정 문장:
  - “지금 당장 없어도 큰 불편이 없다.”
  - “구매 이유가 ‘기분전환/보상/세일 압박’ 중심이다.”
- 합리 신호 판정 문장:
  - “구체적인 문제를 해결하기 위한 구매다(언제/어디서/무엇을).”
  - “구매를 미루면 실제 손해/불편이 발생한다.”

채점(1~5):
1점: ‘갖고 싶다’ 중심 + 미뤄도 영향 거의 없음
2점: 필요가 애매하며 감정/세일 요인이 강함
3점: 필요는 있으나 긴급도 낮음(미뤄도 됨)
4점: 필요가 명확하고 미루면 불편이 큼
5점: 필수에 가까움(업무/학업/건강/안전 등) + 대체 어려움

---

### 3단계: 가치/효율(가격 대비 사용) 평가
- 충동 신호 판정 문장:
  - “사용 빈도가 낮고 1회 사용당 비용이 높다.”
  - “대체재(이미 가진 것/저렴한 대안)로 충분하다.”
- 합리 신호 판정 문장:
  - “사용 빈도가 높고 1회 비용이 납득된다.”
  - “대체재 대비 효용이 확실히 크다(시간/품질/편의).”

채점(1~5):
1점: 저빈도/고비용/대체재 충분 → 가성비 최악
2점: 사용 빈도 불확실 + 대체 가능성 높음
3점: 무난하나 가격 대비 효용이 압도적이진 않음
4점: 사용 빈도 높고 비용/효용 균형 양호
5점: 고빈도 + 절약/효율 개선이 확실(시간/품질/지출 절감)

---

### 4단계: 리스크/후회 확률 평가
- 충동 신호 판정 문장:
  - “사이즈/호환/취향 실패 가능성이 크다.”
  - “반품/AS가 까다롭거나 추가 비용이 숨겨져 있다.”
  - “구매 시점이 감정적 상태(스트레스/피로/보상심리)다.”
- 합리 신호 판정 문장:
  - “실패 가능성이 낮고(호환/사이즈 확정), 반품/AS가 안전하다.”
  - “구매 결정이 감정이 아니라 목적/데이터 기반이다.”

채점(1~5):
1점: 실패 확률 매우 높음 + 반품/AS 불리 + 감정구매 신호 강함
2점: 리스크 높고 불확실 요소가 큼
3점: 리스크 중간(반품 가능/정보 일부 확인)
4점: 리스크 낮음(조건 명확, 실패 대비 가능)
5점: 리스크 매우 낮음(검증된 선택, 반품/AS 안전, 숨은 비용 낮음)

---

### 5단계: 행동 처방/구매 조건 최적화
- 충동 신호 판정 문장:
  - “지금 당장 결제하지 않으면 안 될 이유가 없다.”
  - “48시간만 지나도 구매 욕구가 크게 줄 가능성이 높다.”
- 합리 신호 판정 문장:
  - “구매해도 되는 조건(예산/최저가/필수 옵션)이 명확하다.”
  - “보류 시 대기 규칙(48시간 룰/가격알림/대안 탐색)이 실행 가능하다.”

채점(1~5):
1점: 즉시 결제 충동이 강하고 통제 장치 없음
2점: 보류 규칙이 약하거나 실행 계획이 없음
3점: 조건부로 가능(보류/대안 탐색 필요)
4점: 구매 조건 최적화 가능(옵션/가격/예산 준수)
5점: 구매/보류 모두 실행 플랜이 구체적이며 후회 방지 장치가 있음

---

## 출력 형식(반드시 준수)
결론: (구매 권장 / 보류 / 비구매 권장)
총점: (0~25)
단계별 판정 및 점수:
1단계 판정: ...
1단계 점수: 1~5
2단계 판정: ...
2단계 점수: 1~5
3단계 판정: ...
3단계 점수: 1~5
4단계 판정: ...
4단계 점수: 1~5
5단계 판정: ...
5단계 점수: 1~5
핵심 근거 요약: (3~6줄)
행동 처방: (오늘 할 1~3가지)
추가 질문: (정보가 부족한 경우에만 2~5개)
'''

product_url = "https://www.coupang.com/vp/products/6396591736?itemId=13660776260&vendorItemId=80329308161&src=1042501&spec=10304968&addtag=400&ctag=6396591736&lptag=I13660776260&itime=20260119152147&pageType=PRODUCT&pageValue=6396591736&wPcid=17685438160070524287168&wRef=search.naver.com&wTime=20260119152147&redirect=landing&mcid=8471cc6e038d4a6daed4b23d93acacc6&n_ad_group=grp-a001-02-000000041144746&n_ad=nad-a001-02-000000441906598&n_rank=1&n_media=27758&n_query=%EC%8B%A4%EB%B0%94%EB%8B%88%EC%95%88%EC%BD%94%EB%81%BC%EB%A6%AC"
user_context = """예산: 3만원
구매 이유: 힐링하려고
예상 사용 빈도: 상시
유사 제품 보유: 없음
반품 조건: 아직 확인 못함
"""

user_prompt = build_user_prompt_from_url(product_url, user_context=user_context)

res = client.responses.create(
    model="gpt-4o-mini",
    input=[
        msg("system", system_prompt),
        msg("user", user_prompt)
    ],
    temperature=0.3
)

print(res.output_text)
````
