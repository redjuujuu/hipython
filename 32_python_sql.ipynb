```python
#sql 쿼리 실행
sql = """
select 담당자명
	, 고객회사명
	, 마일리지
	, 고객.도시    -- 선생님이 그냥 도시라고 썼었는데 모호하다고 해서 더 명확하게함
	, 도시_평균마일리지
	, 도시_평균마일리지 - 마일리지 AS 차이

FROM 고객
	,(Select 도시
			,AVG(마일리지) AS 도시_평균마일리지
	 From 고객
     Group by 도시
     ) AS 도시별요약
where 고객.도시 = 도시별요약.도시;
"""
sql
# 쿼리만 바꿔주면 됨
```

```python
def getconnection():
    con = pymysql.connect(
        host='localhost',
        user='root',
        password='1234',
        db='wntrade',
        charset='utf8'
        ) #한글처리 (charset = 'utf8')
    return con
```

```python
import pandas as pd
```

```python
#my sql 연결함수 호출
con = getconnection()

try:
    #2. 커서 생성  -> 커서 생성코드가 with문에 있다는 것 주의하기
    with con.cursor() as cursor:
        #3. sql 실행
        cursor. execute(sql)       #커서가 유용할때만 사용해
        #4. 쿼리실행결과 받기
        result = cursor.fetchall()
        #dataframe으로 결과 반환하기
        result_df = pd.DataFrame(result)

finally:
    #6. 연결 종료
    con.close()
```
