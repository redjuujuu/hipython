# 딕셔너리

```python
#키는 유니크해야 함 - 겉으로 보기에 비슷해 보여도  --> mutable임 바뀔 수 있음
```

```python
person = {"이름" : 'alice' , 'age' : 25 , 'city' :'서울' }
person_1 = {"이름" : 'bob' , 'age' : 30 , 'city' :'부산' }
```

```python
person["이름"]  #인덱싱이나 슬라이싱은 다 대괄호로 한다고 통일
```

```python
person['age']
```

```python
person['city'], person_1['city']
```

```python
for per in person.values():     #저 위에  콜론구조 안맞으면 어떡해?
    print(per) #key:value 1쌍
```

```python
for per, val in person.items():  #items는 뭐야
    print(per, val)

# items()는 딕셔너리에서 키와 value(값)을 한쌍씩 꺼내주는 메서드임
```

```python
person.get('직업')   #get??
```

```python
person.get('age')
```

```python
print(person.keys())
```

```python
print(person.values())
```

```python
print(person.items())
```

```python
print(person.pop('age'))
```

```python
person    #value 값이 빠지면 key값도 사라지는구나!!  - list와의 차이점
```

```python
person.update({'age' : 26, '직업' : '학생'})
print(person)
```

```python
job_cur_status = person.get('job', '정보없음') #기본값 설정하는 것
job_cur_status
```

```python
age_cur_status = person.get('age', '정보없음') #기본값 설정하는 것
age_cur_status
```

```python
if job_cur_status == '정보없음' :
    new_job = input('직업을 입력해주세요')
    person.update({'job': new_job})
    print(person.get('job'))
```
