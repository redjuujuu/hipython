```python
#메뉴 삭제
def remove_snackbox(ssbox):
    make_msg = input('빼고 싶은 간식을 선택하세요.')
    ssbox.remove(make_msg)   #remove 자리에 뭘 대신 써야할 지 모르겠어
    return ssbox
```

```python
def order_confirm():              # 얘는 받아야 할지 아닐지 - 얘는 받을 필요 없음... display_snackbox를 거치고 와야하기 때문에
    make_msg = input('주문을 확정하시겠습니까? (y/n)')  # 얘 또 써도 상관없음  - yes이면 order 확정... 
    if make_msg == 'y':   #y,n으로 바꾼이유 : 쥬피터 노트북에 no 쳤을 때 o에 무슨 기능이 있어서...
        return True
    else:
        return False                 #yes no로 받으면 문자열 boolean으로 받으면 ....
    
```
