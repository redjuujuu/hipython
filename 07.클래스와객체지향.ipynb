# 자동차 클래스 정의
- self.속성명 : 인스턴스 속성
- 속성명 : 클래스 속성

- __init__(self) 생성자, 매개변수 => 인스턴스 속성 설정
- def 메소드(self): 메소드 정의
- 정적 메소드 : 유틸리티
- 클래스 메소드(cls) : 클래스 속성 관리, @classmethod "데코레이터"

```python
class Car: 
    #1 = 클래스 속성
    instance_cnt = 0

   # def __init__(self):
        #print('자동차 객체가 생성되었습니다.')

    def __init__(self, color=''): #생성자 정의
        #2 - 객체의 속성
        self.color = color

        Car.instance_cnt += 1
        print(f'{self.color} 자동차 객체가 생성되었습니다.')

    def auto_pilot(self):  #self만 주면됨
        print(f'{self.color} 자동차가 자율주행중입니다.')
        
        print(f'{Car.instance_cnt}대의 자동차가 운행중입니다.')
        #00대의 자동차가 운행중입니다.
    
    @classmethod
    def count_check(cls):
        print(f'자동차 대수 : {cls.instance_cnt}')

    @staticmethod   #이건 클래스에 넣을필요도 없다 - 이거 무슨말이야
    def type_check(type_code):
        if type_code > 20 :
            print('전기자동차')
        elif type_code < 20 :
            print('내연자동차')
        else :
            print('하이브리드')
```
