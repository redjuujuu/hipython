```python
class Bicycle:  #python Object가 원조부모
    def __init__(self, color, brand):    #왜 self하나가 들어갔는데 해결이 됐지?
        self.color = color
        #내가가지고있는컬러 위에 컬러는 받은 컬러
        self.brand = brand

        
    def move(self):
        print(f'{self.color} {self.brand} 자전거가 움직입니다.')   #여기선 왜 self.color??
```

```python
#DataFrame()  #객체생성 - 클래스에 괄호붙이면 됨

Bicycle()
```

```python
#객체 생성
# b1 = Bicycle()
# b2 = Bicycle()

b1 = Bicycle('red', '자이언트') #객체생성
b2 = Bicycle('white', '캄파놀로')
```

```python
b1.move()   #객체 출력시 이렇게 나왔었음
```

```python
b2.move()
```
