5. 어떤 숫자 x가 주어졌을때, 그 숫자가 짝수인지 홀수인지 계산하는 코드를 작성하세요.

```python
#005
#2로 나누었을 때 나머지가 0이면 짝수, 1이면 홀수
```

```python
#005 answer
x =int(input('홀짝을 판별할 숫자를 입력하세요'))
나머지 = x % 2   
if 나머지 == 1 :    #비교 == 같은가?
    print(f'{x}는 홀수입니다.')
else :
    print(f'{x}는 짝수입니다.')
```

왜 나누기가 아니라 %로 쓰지??? 
=> 홀짝을 판단하려면 몫이 아니라 나머지가 필요하기 때문에 나눗셈(/)이 아니라
나머지 연산자(%)를 쓴다.
