```python
data = list(range(-100,100))
print(len(data))
```

```python
data
```

```python
#리스트에 양수만 남긴다 : 목표
for x in data :
    if x < 0 :
        data.remove(x)  #원본데이터가 줄어들고 있는데... 원래 알던데로 가니까 지금 음수는 짝수만 빠짐

        #이런경우 오류가 남... 누적 후 한꺼번에 처리하는게 나음 새로운 리스트 만들어서 조건 반대로 쓰기
        #조건에 맞는 애들만 골라서 추가추가하기
        #원본에서 하되 내 조건에 맞는 애들을 고르는게 나음!!!

filtered = []
for x in data:
    if x >= 0 :
        filtered.append(x)
```
